## Advertising Project

In this project, I will be using data related to a fictional advertising company. The goal will be to understand whether or not a particular internet user clicked on the company's advertisement. 

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

## Logistic Regression

The data will be split into training and test sets. Preparation will be performed on the relevant features if necessary (imputation, encoding, etc.)

```python
#importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
#feature columns data frame 
X = df[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
#target column dataframe
y = df['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
#instantiate Logistic Regression Model
lr = LogisticRegression(solver='liblinear')

#train model
lr.fit(X_train, y_train)
```

## Predictions + Evaluation

```python
#predict whether a given user clicked on the ad in the test data
predictions = lr.predict(X_test)
```

#### Classification Report

This report will contain the precision, recall, f1-score, and support
- Documention on each: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_recall_fscore_support.html#sklearn.metrics.precision_recall_fscore_support

```python
#importing classification report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
